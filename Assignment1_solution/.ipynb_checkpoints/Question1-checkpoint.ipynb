```python
import nltk
```

```python
# files under Package Gutenburg are:
nltk.corpus.gutenberg.fileids()
```

```python
a= nltk.corpus.gutenberg.words( 'austen-persuasion.txt')
```

```python
# Number of words in the file 'austen-persuasion.txt' are:
len(a)
```

```python
set(a)
```

```python
# Number of unique words in the file are:
len(set(a))
```

```python
    words= set(a)
    words
    
```

```python
# To get the unique words from our file a
uniques=[]
for word in a:
    if word not in uniques:
        uniques.append(word)
# To calculate the frequency/count of unique words       
counts= []
for unique in uniques:
    count=0
    for word in a:
        if word==unique:
            count+=1
    counts.append((count, unique))
    
# To get the top 10 frequency of words
counts.sort() # This will sort all frequencies in increasing order
counts.reverse() # This will arrange list in decreasing order of frequencies
# For printing the top 10 words with highest frequency:
for i in range(min(10, len(counts))):
    count, word = counts[i]
    print('%s %d' % (word, count))
```
