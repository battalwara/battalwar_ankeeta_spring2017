# Question 2 Part 1:

Read all json files under folder data. Create a heirarchy of folder structures to store all the files. 
Folder= DataProcessed 

```python
import glob
import json
import os
```

```python
# To get the different categories/terms from all json files
list_terms = [];                                           
for filejson in glob.glob('data/*.json'):                         # Looking through all the json files in data folder
    with open(filejson) as f:                               
        json_data = json.load(f)                            
        if json_data['term'] not in list_terms:             
            list_terms.append(json_data['term'])             
print(list_terms)          
```

```python
# Function to create directory
def makeDirectory(path):                                         
    if not os.path.exists(path):                             
        os.makedirs(path)                                   
        print(path+" was created successfully")
    else:
        print(path+" directory already exists")                        
directoryStructure_path = 'DataProcessed'           
makeDirectory(directoryStructure_path)   
```

```python
# To create subfolders in the base directory of all the countries present as location in all json files
countries=[]                                                 
for filejson in glob.glob('data/*.json'):                         
     with open(filejson) as f:                               
        json_data = json.load(f)                             
        location = json_data['location']                     
        if location['country'] not in countries:             
            countries.append(location['country'])
            makeDirectory(directoryStructure_path+'/'+location['country'])
print(countries)     
```

```python
for i in range(0,len(countries)):                           
    states=[];                                               
    for filejson in glob.glob('data/*.json'):
        with open(filejson) as f:                               
            json_data = json.load(f)
            location = json_data['location']
            if location['country']==countries[i]:
                 if location['state'] not in states:    
                    states.append(location['state'])  #This will find the unique states for a coutry & store them in a list
    for state in states:                                     #Creating subdirectories of states under the countries using our function        
        makeDirectory(directoryStructure_path+'/'+countries[i]+'/'+state)
    for j in range(0,len(states)):
        cities=[];                                          
        for filejson in glob.glob('data/*.json'):
            with open(filejson) as f:                               
                json_data = json.load(f)
                location = json_data['location']
                if location['country']==countries[i]:
                    if location['state']==states[j]:
                        if location['city'] not in cities:   
                            cities.append(location['city']) #This will find the unique cities for a state & store them in a list
        
        for city in cities:                                 #Creating subdirectories of cities and terms under the city using our function 
            makeDirectory(directoryStructure_path+'/'+countries[i]+'/'+states[j]+'/'+city)
            makeDirectory(directoryStructure_path+'/'+countries[i]+'/'+states[j]+'/'+city+'/'+list_terms[0])
            makeDirectory(directoryStructure_path+'/'+countries[i]+'/'+states[j]+'/'+city+'/'+list_terms[1])
            makeDirectory(directoryStructure_path+'/'+countries[i]+'/'+states[j]+'/'+city+'/'+list_terms[2])
```

```python
# To copy the json files into the respective locations, using copyfile from shutil package
from shutil import copyfile
for filejson in glob.glob('data/*.json'):
        with open(filejson) as f:    # to get all attributes for comparison from the selected json file                           
            json_data = json.load(f)
            _location = json_data['location']
            _country = json_data['location']['country']
            _state = json_data['location']['state']
            _city = json_data['location']['city']
            _term = json_data['term']
            if not os.path.exists('DataProcessed/'+_country+'/'+_state+'/'+_city+'/'+_term):
                os.makedirs('DataProcessed/'+_country+'/'+_state+'/'+_city+'/'+_term)
            copyfile('Data/'+filejson[5:],'DataProcessed/'+_country+'/'+_state+'/'+_city+'/'+_term+'/'+filejson[5:])

           
```
