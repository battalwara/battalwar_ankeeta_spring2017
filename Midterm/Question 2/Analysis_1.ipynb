# Analysis 1

- In this analysis, I've taken all the archive data for 3 months of 2016.
- With this data, I've analyzed which countries amongst those that U.S. does business actively; Inida, Japan, China, France, are most frequently mentioned in the headlines.

```python
import os
import json
import matplotlib.pyplot as matplot

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter

key1= "india"
key2= "japan"
key3= "china"
key4= "france"

count_India=0
count_China=0
count_Japan=0
count_France=0
```

```python
for filename in glob('ArchiveData/*.json'):   

    with open(filename) as file:                                
        data = json.load(file)                             
        response=data['response']                 
        docs=response['docs']                      
        for doc in docs:  
            wordlist =(doc['headline']['main'])
            words= word_tokenize(wordlist)
            
            india_words = [word  for word in words if word in key1]    
            for key1 in india_words:
                count_India+=1
                
            japan_words = [word  for word in words if word in key2]    
            for key2 in japan_words:
                count_Japan+=1
                
            china_words = [word  for word in words if word in key3]    
            for key3 in china_words:
                count_China+=1
                              
            france_words = [word  for word in words if word in key4]    
            for key4 in france_words:
                count_France+=1
```

```python
print("Number of times India appeared in the headlines is:")
print(count_India)
print("Number of times China appeared in the headlines is:")
print(count_China)
print("Number of times Japan appeared in the headlines is:")
print(count_Japan)
print("Number of times France appeared in the headlines is:")
print(count_France)
```

```python
import matplotlib.pyplot as plot
labels=[]
sizes=[]
colors = ['red','yellow','blue','green']

labels=('India', 'China', 'Japan', 'France')
sizes= (count_India,count_China,count_Japan,count_France )
plot.title('Country-wise frequency distribution in headlines',y=1.18)
plot.pie(sizes,colors=colors,labels=labels,radius=1.0)
plot.axis('equal')
plot.show()
```
